```python
import pandas as pd

# Step 1: Load the data from the JSON file and show products
products = pd.read_json('../cleaning/products1.json')
# Display the first few rows of the DataFrame
products.head()
```

```python
products.shape
```

```python
most_popular = products.sort_values('rating', ascending=False)
most_popular.head(8)
```

```python
# Count the number of products for each rating
rating_counts = products['rating'].value_counts().sort_index()

# Visualize the data as a bar chart
plt.figure(figsize=(8, 6))
rating_counts.plot(kind='bar', color='skyblue')
plt.title('Number of Products by Rating', fontsize=14)
plt.xlabel('Rating', fontsize=12)
plt.ylabel('Number of Products', fontsize=12)
plt.xticks(rotation=0)  # To make sure the X-axis labels are horizontal
plt.show()
```

```python
# Count the number of products for each category
category_counts = products['category'].value_counts()

# Visualize the data as a bar chart
plt.figure(figsize=(10, 6))
category_counts.plot(kind='bar', color='lightcoral')
plt.title('Number of Products per Category', fontsize=14)
plt.xlabel('Category', fontsize=12)
plt.ylabel('Number of Products', fontsize=12)
plt.xticks(rotation=45, ha='right')  # Rotate X-axis labels for better readability
plt.tight_layout()  # Adjust layout to avoid clipping of labels
plt.show()
```
